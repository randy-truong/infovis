```python
import pandas as pd
import numpy as np
```

```python
# Read in data
data = pd.read_csv('/content/drive/MyDrive/cs485/20162020allstatesO3.csv')
data
```

```python
# Date initially string, convert to datetime
data['Date'] = pd.to_datetime(data['Date'])
# Date convert datetime to year-month, no day
data['Date'] = pd.to_datetime(data['Date'],format='%m%Y', errors='coerce').dt.to_period('m')
```

```python
data
```

```python
data.columns
```

```python
# Drop unnecessary columns
cleaner_data = data.drop(['Source', 'Site ID', 'Site Name', 'STATE_CODE', 'CBSA_CODE', 'CBSA_NAME', 'COUNTY_CODE', 'COUNTY', 'SITE_LATITUDE', 'SITE_LONGITUDE', 'AQS_PARAMETER_CODE', 'AQS_PARAMETER_DESC'], axis=1)
cleaner_data
```

```python
# Group states by date, then perform aggregations 
cleaner_sum = cleaner_data.groupby(['Date','STATE']).sum()
cleaner_sum
```

```python
cleaner_mean = cleaner_data.groupby(['Date','STATE']).mean()
cleaner_mean
```

```python
test = pd.read_csv('/content/drive/MyDrive/cs485/2016-20O3_clean_mean.csv')
test
```

```python
dates = np.unique(test['Date'])
keys = []

for i,j in enumerate(dates):
  for k in test['Date']:
    if j == k:
      keys.append(i)

print(keys)

```

```python
test['Date_Keys'] = keys
test
```

```python
cleaner_data_keys = test.drop(['Unnamed: 0', 'V1', 'POC', 'DAILY_OBS_COUNT', 'PERCENT_COMPLETE'], axis=1)
cleaner_data_keys

cleaner_data_keys['DM8O3_Concentration'] = cleaner_data_keys['Daily Max 8-hour Ozone Concentration']
cleaner_data_keys = cleaner_data_keys.drop(['Daily Max 8-hour Ozone Concentration'], axis=1)
cleaner_data_keys
```

```python
print(max(cleaner_data_keys['DAILY_AQI_VALUE']))
print(min(cleaner_data_keys['DAILY_AQI_VALUE']))
print(np.unique(cleaner_data_keys['STATE']))
print(cleaner_data_keys['STATE'][0])
```

```python
# Fix 'District Of Columbia'
states = []
for i, j in enumerate(cleaner_data_keys['STATE']):
  
  if j == 'District Of Columbia':
    states.append('District of Columbia')
  else:
    states.append(j)

print(states)
print(len(states))
```

```python
cleaner_data_keys['State_Fixed'] = states
```
