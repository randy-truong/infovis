```python
import pandas as pd
import numpy as np
```

```python
# Read in data
data = pd.read_csv('/content/drive/MyDrive/cs485/20162020allstatesCO.csv')
data
```

```python
# Date initially string, convert to datetime
data['Date'] = pd.to_datetime(data['Date'])
# Date convert datetime to year-month, no day
data['Date'] = pd.to_datetime(data['Date'],format='%m%Y', errors='coerce').dt.to_period('m')
```

```python
data
```

```python
data.columns
```

```python
# Drop unnecessary columns
cleaner_data = data.drop(['Source', 'Site ID', 'Site Name', 'CBSA_CODE', 'CBSA_NAME', 'COUNTY_CODE', 'COUNTY', 'SITE_LATITUDE', 'SITE_LONGITUDE'], axis=1)
cleaner_data
```

```python
# Group states by date, then perform aggregations 
cleaner_sum = cleaner_data.groupby(['Date','STATE']).sum()
cleaner_sum
```

```python
cleaner_sum.head(52)
```

```python
cleaner_mean = cleaner_data.groupby(['Date','STATE']).mean()
cleaner_mean
```
